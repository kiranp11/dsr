```python
from urllib.request import urlopen

with urllib.request.urlopen('http://www.digbib.org/Johann_Wolfgang_von_Goethe_1749/Faust_I?textonly=1') as response:
   text = response.read()
```

```python
print(text.decode("utf-8", errors = "ignore"))
```

```python

tokens = text.decode("utf-8", errors = "ignore").split(' ')
vocab = {'und'}
for token in tokens:
    vocab.add(token.lower())

print(vocab)
```

```python
import re

tokens = re.split('\W', text.decode("utf-8", errors = "ignore"))
vocab = {'und'}
for token in tokens:
    vocab.add(token.lower())

print(sorted(vocab))
```

```python
import re
from urllib.request import urlopen

with urllib.request.urlopen('http://www.digbib.org/Johann_Wolfgang_von_Goethe_1749/Faust_I?textonly=1') as response:
   text = response.read()

tokens = re.split('\W', text.decode("utf-8", errors = "ignore"))

print(sorted(set(word.lower() for word in tokens if word.isalpha())))
```
