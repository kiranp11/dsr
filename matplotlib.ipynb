```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from pylab import *
```

```python
x = np.linspace(0, 5, 10)
```

```python
y = x ** 2
```

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title goes here')
show()
```

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y, x, 'g*-');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes


# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

```
