```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(random_state=42)
```

```python
X.shape
```

```python
# Clustering
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, n_jobs=-1, random_state=42)
kmeans.fit(X)
```

```python
cluster_labels = kmeans.predict(X)
```

```python
print(cluster_labels)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.title("K-Means of moons for K = 2")
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, n_jobs=-1)
cluster_labels_db = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels_db)
plt.title("DBScan of Moons")
```
